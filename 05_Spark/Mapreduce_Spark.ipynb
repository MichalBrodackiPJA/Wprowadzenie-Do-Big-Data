```python
import os
os.environ["JAVA_HOME"] = "/usr/lib/jvm/java-11-openjdk-amd64"
os.environ["SPARK_HOME"] = "./spark-3.5.0-bin-hadoop3"
```

```python
import findspark

findspark.init()
findspark.find()
```

```python
from pyspark import SparkConf, SparkContext


conf = SparkConf().setMaster("local").setAppName("CustomerOrders")
sc=SparkContext(conf=conf)
```

```python
def parse_line(line):
  ID, _, value = line.split(',')
  return ID, float(value)

lines = sc.textFile('customer-orders.csv')
pairs = lines.map(parse_line)
vals_per_item = pairs.reduceByKey(lambda x,y: x+y)
results = vals_per_item.collect()

for result in results:
  print(result)
```

```python
def get_value(line):
  return line[1]

lines = sc.textFile('customer-orders.csv')
pairs = lines.map(parse_line)
vals_per_item = pairs.reduceByKey(lambda x,y: x+y)
vals_per_item_sorted = vals_per_item.sortBy(get_value, ascending=True)
results = vals_per_item_sorted.collect()

for result in results:
  print(result)
```
